```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
def draw_line_for_two_points(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef*mi - intercept)/coef), mi],[((-coef*ma - intercept)/coef), ma]])
    plt.plot(points[:,0], points[:,1])
    
    
```

```python
# the values (-10, 15) are. range of values on x-axis
def draw_line(coef,intercept, mi, ma):
    xx=np.linspace(-10,15)
    y= coef * xx +intercept
    plt.plot(xx,y)
```

```python


regularizations_alpha=[0.0001, 1, 100]
outliers = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
for i,each_regularizer in enumerate(regularizations_alpha):
    plt.figure(figsize=(20,5))
    # Doing initilazing the X & Y values here
    X= b * np.sin(phi)
    Y= a * np.cos(phi)
    for j,each_outlier in enumerate(outliers):
        plt.subplot(i+1, 5, j+1)
        #append outliers in X
        X = np.append(X,each_outlier[0]).reshape(-1,1)
        #append outliers in Y
        Y = np.append(Y,each_outlier[1]).reshape(-1,1)
        model = SGDRegressor(alpha=each_regularizer, eta0=0.001, learning_rate='constant',random_state=0)
        model.fit(X,Y)
        coef = model.coef_
        intercept = model.intercept_
        mi = np.min(X)
        ma = np.max(Y)
#         print(coef,intercept, mi, ma)
        draw_line(coef,intercept, mi, ma)
        plt.scatter(X,Y)
    plt.show()
```
